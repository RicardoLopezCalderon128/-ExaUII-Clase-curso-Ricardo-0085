1. Definición del Problema
El objetivo es diseñar una estructura de datos utilizando el paradigma de Programación Orientada a Objetos para gestionar la información de cursos académicos. Se requiere la creación de un modelo (clase) que permita capturar y visualizar datos de manera organizada.

![texto del vínculo](https://raw.githubusercontent.com/RicardoLopezCalderon128/-ExaUII-Clase-curso-Ricardo-0085/refs/heads/main/Captura%20info%20sisi.JPG)

```python
class Curso:
    def __init__(self, tematica, duracion, profesor):
        self.tematica = tematica
        self.duracion = duracion
        self.profesor = profesor

    def leerDatos(self):
        self.tematica = input("Ingrese la temática del curso: ")
        self.duracion = input("Ingrese la duración del curso: ")
        self.profesor = input("Ingrese el nombre del profesor: ")

    def mostrarDatos(self):
        print("Temática:", self.tematica)
        print("Duración:", self.duracion)
        print("Profesor:", self.profesor)

# Crear dos objetos de la clase Curso
curso1 = Curso("", "", "")  # Inicialmente vacíos
curso2 = Curso("", "", "")  # Inicialmente vacíos

# Leer datos interactivos para el curso1
print("Ingresar datos para el curso 1:")
curso1.leerDatos()

# Mostrar datos para el curso 1
print("\nDatos del curso 1:")
curso1.mostrarDatos()

# Leer datos interactivos para el curso 2
print("\nIngresar datos para el curso 2:")
curso2.leerDatos()

# Mostrar datos para el curso 2
print("\nDatos del curso 2:")
curso2.mostrarDatos()

print("programa echo por Lopez Ricardo NL 29")
```
